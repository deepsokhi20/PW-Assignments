App routing is the mapping of URLs to a specific funtion that will handle the logic for the URL. Modern web frameworks use more meaningful URls to help users remember the URls and make navigation simpler. To bind a function to an URL path we use @app.route("/"). for example, while printing hello world in question 2 , we have written the code and in that code if we wrote @app.route("/hello")  then the URL for that web page will be "https://white-fireman-dryci.pwskills.app:5000/hello"
we use app routes to link our funtion with the url which makes it easier to remember for the developers.
