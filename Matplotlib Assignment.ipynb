#Q1 
Matplotlib is a python library which is used to create visuals of large data to gain insights from it. It helps users to make different bar graphs, piecharts to understand the data well.
the five plots that can be plotted using Pyplot module of matplotlib are:-
1. line plot- a simple plot where data points are joined by straight line useful for visualizing trends and relationships between variables.
2. scatter plots- A plot where individual data points are represented as dots on the chart, useful for visualizing the distribution and correlations between two variables
3. Bar plot- a plot which uses rectangular bars to represent data values, useful for showing data into grouped catagories.
4. Histogram- A graphical representation of the distribution of numerical data, showing how many data points fall under the same category.
5. Pie chart- A circular chart divided into sectors, each sector represents a portion of the whole data, used to show data in percentage.

scatter plot is a plot where individual data points are represented as dots on the chart useful for visualing trends amd relationships between variables

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0,2,50)
y = 3 + np.random.normal(0,2,len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel("X axis")
plt.ylabel("Y label")
plt.title("scatter plot")
```

#Q3 

the subplot() function in matplotlib is used to create multiple plots in the same figure.
it is advantageous when you have to compare different plots of the same data side by side. the subplot() function takes three integer arguments"num_rows", "num_cols", and "plot_number" which represent the number of rows and number of columns and the position of the subplot within the grid, respectively.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0,1,2,3,4,5])

y1= np.array([0,100,200,300,400,500])
y2 = np.array([50,20,40,20, 60, 70])
y3 = np.array([10,20,30,40,50,60])
y4 = np.array([200,350,250,550,450,150])

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("Line plot 1")

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("Line plot 2")

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("Line plot 3")

plt.subplot(2,2,4)
plt.plot(x, y4)
plt.title("Line plot 4")

plt.tight_layout()
plt.show
```

#Q4 

A bar plot is type of chart that uses a rectangular bars to represent data values. the lenght of each bar represents the value. the x axis usually represents the categories and y axis represents the numerical data.
the bar plots are used to visualize data which helps to understand the data well. it can be used to compare categorical data which makes it easy to identify which category has the highest or lowest value.

```python
company= np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company, profit)
plt.xlabel("COMPANY")
plt.ylabel("PROFIT")
plt.show()
```

```python
plt.barh(company, profit)
plt.xlabel("COMPANY")
plt.ylabel("PROFIT")
plt.show()
```

A box plot is also known as a box and whisker plot, is a statistical visualization tool that provieds a concise summary of the distribution of a dataset. It displays the five number summary of a datasest, which includes the minimum, first quartile Q1, median second quartile Q2, third quartile Q3 and the maximum. the box plot consists of a rectangular box that spans from the first quartile to the third quartile, with a vertical line at the median, "whiskers" extend from the box to the minimum and maximum data points that fall within certain distance from the quartiles. any data points beyond the whiskers are considered outliers and are typically represented as individual points.
box plot helps to visualize the spread, skewness, presence of outliers in a dataset.
they help in comparison of multiple datasets side by side, which makes it easier to see distances in their central tendency and variability.
